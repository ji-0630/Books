### 서울시 관서별 5대 범죄 현황
#### 1. pandas 이용하여 데이터 정리하기

```python
import numpy as np
import pandas as pd

crime_anal_police = pd.read_excel("/content/data/crime_in_seoul.xlsx")
crime_anal_police.head()
```

```python
import googlemaps

gmaps = googlemaps.Client(key=gmaps_key)
```

```python
gmaps.geocode('서울중부경찰서', language ='ko')
```

```python
station_name = []

for name in crime_anal_police['관서명']:
  station_name.append('서울' + str(name[:-1]) + '경찰서')

station_name
```

```python
station_addreess = []
station_lat = []
station_lng = []

for name in station_name:
  tmp = gmaps.geocode(name, language = 'ko')
  station_addreess.append(tmp[0].get("formatted_address"))

  tmp_loc = tmp[0].get("geometry")
  station_lat.append(tmp_loc['location']['lat'])
  station_lng.append(tmp_loc['location']['lng'])
  print(name + "-->" + tmp[0].get("formatted_address"))
```

```python
gu_name = []
for name in station_addreess:
  tmp = name.split()
  tmp_gu = [gu for gu in tmp if gu[-1] == "구"][0]
  gu_name.append(tmp_gu)

crime_anal_police['구별'] = gu_name
crime_anal_police.head()
```

```python
##금천경찰서는 관악구에 있어서 예외처리 해야 함.

crime_anal_police.loc[crime_anal_police['관서명'] == "금천서", ['구별']] == '금천서'
crime_anal_police[crime_anal_police['관서명']=='금천서']
```

### 2.pivot_table 이용해서 데이터 정리하기

```python
crime_anal = pd.pivot_table(crime_anal_police, index = "구별", aggfunc=np.sum)
crime_anal.head()
```

```python
crime_anal['강간검거율'] = crime_anal['강간 검거']/crime_anal['강간 발생']*100
crime_anal['강도검거율'] = crime_anal['강도 검거']/crime_anal['강도 발생']*100
crime_anal['살인검거율'] = crime_anal['살인 검거']/crime_anal['살인 발생']*100
crime_anal['절도검거율'] = crime_anal['절도 검거']/crime_anal['절도 발생']*100
crime_anal['폭력검거율'] = crime_anal['폭력 검거']/crime_anal['폭력 발생']*100

del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']

crime_anal.head()
```

```python
#전년도에 발생한 사건 검거도 포함되므로 검거율 100 넘을 수 있음. 이를 100으로 변경

con_list = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']

for column in con_list:
  crime_anal.loc[crime_anal[column] > 100, column] = 100

crime_anal.head()
```

```python
crime_anal.rename(columns={'강간 발생' : '강간', '강도 발생': '강도', '살인 발생' : '살인', '절도 발생': '절도', '폭력 발생' : '폭력'}, inplace = True)
crime_anal.head()
```

### 3. 데이터 표현을 위해 다듬기

```python
from sklearn import preprocessing

col = ['강간', '강도', '살인', '절도', '폭력']

x = crime_anal[col].values
min_max_scaler = preprocessing.MinMaxScaler() #최소값, 최대값을 이용해 정규화

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_anal_norm = pd.DataFrame(x_scaled, columns = col, index = crime_anal.index)

col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm.head()
```

```python
result_CCTV = pd.read_csv("/content/data/CCTV_result.csv", encoding = "cp949", index_col="구별")


crime_anal_norm[['인구수', 'CCTV']] = result_CCTV[['인구수', '소계']]
crime_anal_norm.head()
```

```python
col = ['강간', '강도', '살인', '절도', '폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal[col], axis=1) #발생건수의 합 구하기
crime_anal_norm.head()
```

```python
col = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm
```

### 4. Seaborn 사용해서 시각화하기

```python
%matplotlib inline 
#notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것 

import matplotlib as mpl 
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

!apt-get update -qq
!apt-get install fonts-nanum* -qq

path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf' 
font_name = fm.FontProperties(fname=path, size=10).get_name()
plt.rc('font', family=font_name)

fm._rebuild()
mpl.rcParams['axes.unicode_minus'] = False
```

```python
sns.pairplot(crime_anal_norm, vars=['강도', '살인', '폭력'], kind='reg', height=3)
plt.show()
```

강도와 폭력, 살인과 폭력, 강도와 살인 모두 양의 상관관계를 보인다.

```python
sns.pairplot(crime_anal_norm, x_vars=['인구수', 'CCTV'], y_vars=['살인', '강도'], kind='reg', height =5)
plt.show()
```

모두 양의 상관관계를 가진다.
특징적인 것은 CCTV가 적은 쪽일 때 많은 살인과 강도 범죄가 일어난다는 것이다.

```python
tmp_max = crime_anal_norm['검거'].max()
crime_anal_norm['검거'] = crime_anal_norm['검거'] / tmp_max*100 #최고항목을 100으로 설정
crime_anal_norm_sort = crime_anal_norm.sort_values(by='검거', ascending = False)
crime_anal_norm_sort.head()
```

```python
target_col = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']

crime_anal_norm_sort = crime_anal_norm.sort_values(by='검거', ascending=False)

plt.figure(figsize=(10,10)) #그래프 사이즈 설정
sns.heatmap(crime_anal_norm_sort[target_col], annot=True, fmt='f', linewidths=.5) #셀값을 표기하고, 데이터타입은 float으로
plt.title('범죄 검거 비율(정규화된 검거의 합으로 정렬)')
plt.show()
```

전반적으로 절도검거율이 매우 낮은 편이며, 살인 검거율이 가장 높다.
도봉구의 검거율은 앞도적으로 높으며, 중구의 검거율이 가장 낮다.

```python
target_col = ['강간', '강도', '살인', '절도', '폭력']
crime_anal_norm['범죄'] = crime_anal_norm['범죄']/5
crime_anal_norm = crime_anal_norm.sort_values(by='범죄', ascending = False)
plt.figure(figsize=(10,10))
sns.heatmap(crime_anal_norm_sort[target_col], annot=True, fmt='f', linewidths=.5)
plt.title('범죄비율 (정규화된 발생 건수로 정렬)')
plt.show()
```

발생 건수로 보았을 때, 강남구가 범죄 발생 건수가 가장 많은 것으로, 도봉구가 가장 적은 것으로 나타났다.
